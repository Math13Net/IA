## comprendre le cycle de création d'une figure

```python
plt.figure(figsize=(4,3))               # dimension en inch
plt.plot(x, y)
```

```python
plt.figure()
plt.plot(x, x**2, c="red", lw=3, ls="--", label="rouge")
plt.plot(x, x**3, c="green", lw=5, ls=":", label="verte")
plt.plot(x, x**0.5, c="black", lw=2, ls = "-", label="noire")
plt.xlabel("temps")
plt.ylabel("tomates")
plt.title("figure des tomates")
plt.legend(loc="upper left")
plt.savefig("figure.png")
plt.show()
```

## organisation des figures : plusieurs graphes sur une même figure

```python
plt.figure()

plt.subplot(2,3,1)
plt.plot(x, x**2, c="red", lw=3, ls="--", label="rouge")
plt.xlabel("temps")
plt.ylabel("tomates")
plt.title("carré")

plt.subplot(2,3,2)
plt.plot(x, x**3, c="green", lw=5, ls=":", label="verte")
plt.xlabel("temps")
plt.ylabel("tomates")
plt.title("cube")

plt.subplot(2,3,6)
plt.plot(x, x**0.5, c="black", lw=2, ls = "-", label="noire")
plt.xlabel("temps")
plt.ylabel("tomates")
plt.title("racine")

```

```python
import numpy as np
x = np.linspace(0, 2, 10)
y = x**2
```

```python
fig, ax = plt.subplots()           # attention, prend un S
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(2, 2, sharex=True)
ax[0,0].plot(x, y)
ax[1,0].plot(x, np.sin(x))
ax[0,1].plot(x, x**2)
ax[1,1].plot(x, x**0.5)
```

## Exercice :
* on a 4 expériences
* imprimer les 4 figures sur un même graphique

```python
dataset = {f"experience {i}" : np.random.randn(100) for i in range(4)}
#dataset.keys()
#dataset.values()
```

```python
def graphique(data):
    n = len(data)
    
    plt.figure(figsize=(6,14))
    
    for k, i in zip(data.keys(), range(1, n+1)):              # créer des tuples de numérotation des clés
        plt.subplot(n,1,i)
        plt.plot(data[k])
        plt.title(k)
        
    plt.show()
```

```python
graphique(dataset)
```
