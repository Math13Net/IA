## 2.3. Fonctions renvoyant une valeur.

Les fonctions précédentes ne font qu'afficher les résultats, et on ne peut pas les réutiliser. 

Pour comprendre ce problème, exécuter la cellule suivante.

````python
def calcul_age(annee_naissance, annee_actuelle):
    age = annee_actuelle - annee_naissance
    print("J’ai", age, "ans cette année")

calcul_age(1987, 2024)
age_de_roberto = age + 8
print("Roberto a", age_de_roberto, "ans cette année")
````

Dans la cellule ci-dessous, la variable `age` n’existe que dans la fonction. On ne peut donc pas l’utiliser.

Pour réutiliser une valeur calculer par une fonction, il faut demander à ce qu’elle soit **renvoyée** par la fonction. 

Pour cela on utilise le mot clef `return`

````python
def calcul_age(annee_naissance, annee_actuelle):
    age = annee_actuelle - annee_naissance
    return age
````

On peut maintenant calculer son age en exécutant la cellule suivante: la valeur renvoyée par la fonction est stockée dans la variable `age_du_prof`.

````python
mon_age = calcul_age(1987, 2024)

# Affichage 
print("J’ai" , mon_age, "ans cette année")

# Et Roberto
age_de_roberto = mon_age + 8
print("Roberto a", age_de_roberto, "ans cette année")
````

---
Une boucle est une partie de code permettant de répéter plusieurs fois une ou plusieurs instructions.
Dans une boucle bornée, on définit le nombre de répétitions.

### Syntaxe

```python
for i in range(a,b):
    instruction_1
    instruction_2
    ...
suite du code
```
### Explications

Les instructions qui sont "dans le bloc", c'est-à-dire dont la ligne commence par une indentation, seront exécutées successivement pour des valeurs de i allant <b><span style="color:orange">de a à b exlu (c-a-d de a à b-1).</span></b>

````python
for i in range(0,10):
    print(i)
````

### Bon à savoir

Lorsqu'on écrit `for i in range(n)`, cela équivaut à `for i in range(0,n)`: le bloc d'instruction(s) sera donc exécuté pour i allant de 0 à n-1 (c'est à dire n fois).
